```python
#data['popularity'].unique().np_percentile(data['popularity'],75)
popularity_percentile75=np.percentile(data['popularity'],75)
popularity_percentile75
```

```python
#data['popularity'].unique().np_percentile(data['popularity'],75)
popularity_percentile25=np.percentile(data['popularity'],25)
popularity_percentile25
```

```python
#data1 = data[data['popularity']< popularity_percentile75]
#data1.shape
```

```python
data1 = data[(data['popularity']< popularity_percentile75) & (data['popularity']> popularity_percentile25)]
data1.shape
```

```python
np.random.seed(14)
random_rows = np.random.choice(list(range(len(data1))), size = 4000,replace=False)
random_rows
```

```python
notsohotlist_sample=(data1.iloc[random_rows, :]).reset_index(drop=True)
notsohotlist_sample
```

```python
notsohotlist=notsohotlist_sample[['artist_name','track_name']]
notsohotlist
```

```python
notsohotlist['track_name'].nunique()
```

```python
not_hot_songs=notsohotlist.drop_duplicates()
```

```python
not_hot_songs['track_name'].nunique()
```

```python
not_hot_songs.columns
```

```python
not_hot_songs.rename(columns = {'track_name' : 'song_title'}, inplace = True)
```
